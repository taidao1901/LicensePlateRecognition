# DATASET

```python
def read_path(path):
  names = os.listdir(path)
  labels = []
  for name in names:
    f = open(os.path.join(path, name), "r")
    for x in f:
      labels.append(x.split()[0])
  return labels
```

```python
def visualize_dataset(data, path=path_kitu):
  tmp = load(os.path.join('/content/drive/MyDrive/CS331_Nhóm/Datasets_Detect', data))
  print('Độ lớn là: ', len(tmp))
  labels = []
  for x in tmp:
    f = open(os.path.join(path, x), 'r')
    for x in f:
      labels.append(x.split()[0])

  c = collections.Counter(labels)
  print(c)
  plt.rcParams["figure.figsize"] = (20,10)
  plt.bar(c.keys(),c.values())
```
