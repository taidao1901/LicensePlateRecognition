```python
def get_name(path_name, txt = True):
  arr_name=[]
  for x in path_name:
    if txt:
      tmp = x.split('.')[0] + '.txt'
    else:
      tmp = x.split('.')[0] + '.jpg'
    arr_name.append(tmp)
  return arr_name
```

```python
names_train = get_name(path_img_train)
names_validation = get_name(path_img_val)
names_test = get_name(path_img_test)
```

```python
def Copy_Img(path,dir):
  for i in path:
    src_path = os.path.join(path_img,i)
    dst_path = os.path.join('/content/drive/MyDrive/Năm3/kỳ2/cs331/CS331_Nhóm/Datasets_Recog/Image/{}'.format(dir),i)
    shutil.copy(src_path, dst_path)
    print('Image {}: {}'.format(dir,i))
```

```python
def Copy_Lb(path,dir):
  for i in path:
    src_path = os.path.join(path_label,i)
    dst_path = os.path.join('/content/drive/MyDrive/Năm3/kỳ2/cs331/CS331_Nhóm/Datasets_Recog/Label/{}'.format(dir),i)
    shutil.copy(src_path, dst_path)
    print('Label {}: {}'.format(dir,i))
```

```python
Copy_Img(path_img_train,'Train')
Copy_Img(path_img_test,'Test')
Copy_Img(path_img_val,'Validation')
Copy_Lb(path_label_train,'Train')
Copy_Lb(path_label_test,'Test')
Copy_Lb(path_label_val,'Validation')
```
